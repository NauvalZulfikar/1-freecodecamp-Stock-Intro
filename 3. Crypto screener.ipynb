```python
#tradingview_ta
def getIntervalRSI(i, indikatorUser, rangeTF):
    i = str(i)
    tf = 0
    
    if rangeTF == "30m":
        tf = Interval.INTERVAL_30_MINUTES
    elif rangeTF == "1HOUR":
        tf = Interval.INTERVAL_1_HOUR
    elif rangeTF == "4HOUR":
        tf = Interval.INTERVAL_4_HOURS
    elif rangeTF == "1DAY":
        tf = Interval.INTERVAL_1_DAY
        
    data = TA_Handler(symbol=i, screener='crypto',
                      exchange='BINANCE',interval=tf)
    
    getRSIStock = (data.get_analysis().indicators)[indikatorUser]
    return getRSIStock

file = "cryptolist.txt"
if os.path.exists(file) and os.path.isfile(file):
    os.remove(file)
    print("Selamat Datang\n")
    
cryptoList = pd.read_json('cryptocurrencies.json', orient = 'index').reset_index().rename(columns={'index':'Symbol',0:'Name'})
cryptoList['pairing_USDT'] = cryptoList['Symbol'] + '-USDT'

# #yfinance
# #Input
# startYear  = input('Enter a starting year  : ')
# startMonth = input('Enter a starting month : ')
# startDay   = input('Enter a starting date  : ')

# #Date
# start = dt.datetime(int(startYear),int(startMonth),int(startDay))
# now = dt.datetime.now()
# df = pdr.get_data_yahoo(cryptoList,start,now)


for i in cryptoList['pairing_USDT']:
    try:
        TF30 = getIntervalRSI(i, 'RSI', '30m')
        TF1H = getIntervalRSI(i, 'RSI', '1HOUR')
        TF4H = getIntervalRSI(i, 'RSI', '4HOUR')

        if (TF30 != None and TF1H != None and TF4H != None):
            if (TF30 <= 50 and TF1H <= 50 and TF4H <= 50):
                print(i + ' is tradable')
                # print('')

            crypto = i
            fields = str(crypto) + '\n'
            
            f = open('cryptolist.txt')
            f.write(fields)
            f.close()
    except Exception:
        continue
```
